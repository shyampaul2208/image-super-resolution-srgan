```python
import torch
import torch.nn as nn
from loss import VGGLoss
from torch.utils.data import DataLoader
from arch import Generator, Discriminator
from dataset import MyDataset
import ssl
import matplotlib.pyplot as plt
ssl._create_default_https_context = ssl._create_unverified_context
import numpy as np
from PIL import Image
from torchvision.utils import save_image
import albumentations as A
from albumentations.pytorch import ToTensorV2
import os
import datetime
from datetime import date, time as ttt
from tqdm import tqdm
```

```python
test_transform = A.Compose([
    A.Normalize(mean=[0, 0, 0], std=[1, 1, 1]),
    ToTensorV2(),
])
```

```python

files = os.listdir("./plot_examples")
gen=Generator()
# Define the path to the saved weights file
#weight_path = "SRResNet_weight.pth"
weight_path="./models/train_gen.pth"
#weight_path="./SRGAN_weight.pth"
# Load the saved weights
saved_G_weight = torch.load(weight_path,map_location=torch.device('cpu'))

# Load the saved weights into the generator network
gen.load_state_dict(saved_G_weight["model_state_dict"])
#gen.load_state_dict(saved_G_weight)

gen.eval()  # Ensure the generator is in evaluation mode


for file in files:
    if file[-3:]=="png":
        image = Image.open(os.path.join("./plot_examples", file))
        print(file)


        # Calculate the new dimensions after downsampling by a factor of 4
        new_width = image.width // 4
        new_height = image.height // 4

        # Perform bicubic downsampling
        image = image.resize((new_width, new_height), Image.BICUBIC)
        lr=image.resize((new_width*4, new_height*4), Image.BICUBIC)

        # Save the downsampled image
        lr.save(f"downsampled_image_bicubic_{file}")

        print(image.size)
        with torch.no_grad():
            # Apply transformations to the image
            transformed_image = test_transform(image=np.asarray(image))["image"]
            transformed_image = transformed_image.unsqueeze(0)

            # Generate high-resolution image from low-resolution input
            upscaled_img = gen(transformed_image)

            # Convert the generated image back to numpy array and save it
            save_image((upscaled_img * 0.5) + 0.5, f"./plot_example_results/gan_{file}")
            print("saved")
```
